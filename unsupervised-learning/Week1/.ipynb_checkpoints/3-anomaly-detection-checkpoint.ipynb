**Anomaly detection**  
**Density estimation**  

 - build a model for probability of x
 - what are the values of the features that have high/low probability
 - ellipses/contours in data set determining how probable new data is

e.g. fraud detection:
 -  Identify unusual users that have p(x) < eps

Widely used

**Gaussian/normal distribution**  

$$ p(x) = \frac{1}{\sqrt{2 \pi \sigma^2}} \exp\left(-\frac{(x - \mu)^2}{2 \sigma^2}\right)$$

$$\mu = \frac{1}{m} \sum_{i=1}^m x^{(i)} $$

$$\sigma = \frac{1}{m} \sum_{i=1}^m (x^{(i)} - \mu)^2 $$

maximum likelihood for mu and sigma  
Can also be over m-1 instead of m - makes little difference

```python
import numpy as np
import matplotlib.pyplot as plt

def gauss(x, mu, sigma):
    return 1/(np.sqrt(2*np.pi*sigma**2)) * np.exp(-((x-mu)**2)/(2*sigma**2))
```

```python
mus = [0,0,0,3]
sigmas = [1,0.5,2,0.5]
x = np.linspace(-6,6,1000)
for mu, sigma in zip(mus,sigmas):
    p = gauss(x,mu,sigma)
    plt.plot(x,p, label = r'$\mu$ = ' + str(mu) + ', $\sigma$ = ' + str(sigma))
    plt.legend()
```
